```python
# SageMaker setup
import sagemaker

role = sagemaker.get_execution_role()
sess = sagemaker.Session(default_bucket='dc-ds-ali-playground')
bucket = sess.default_bucket()
prefix = 'multi-cat-dog'
s3_output_location='s3://{}/{}/{}'.format(bucket, prefix, 'pytorch-multi-model')
```

```python
from sagemaker.pytorch import PyTorch

pytorch_estimator = PyTorch('pytorch-train.py',
                            instance_type='ml.g4dn.xlarge',
                            instance_count=1,
                            framework_version='1.8.0',
                            py_version='py3',
                            role=role,
                            output_path=s3_output_location,
                            hyperparameters = {'epochs': 2, 'batch-size': 4, 'learning-rate': 0.01})
```

```python
pytorch_estimator.fit({'train': 's3://dc-ds-ali-playground/Dog-Cat/train',
                        'val': 's3://dc-ds-ali-playground/Dog-Cat/val'})
```
